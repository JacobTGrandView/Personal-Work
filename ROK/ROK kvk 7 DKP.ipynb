```python
renaming.rename(columns = {'Kills' : 'PrekvkT4kills', 
                           'Unnamed: 5' : 'PrekvkT5kills', 
                           'Kills During KVK' : 'PostkvkT4kills',
                           'Unnamed: 7' : 'PostkvkT5kills', 
                           'Hall of Heroes' : 'T4HOH', 
                           'Unnamed: 9' : 'T5HOH',
                           'Kills.1' : 'T4killsDKP', 
                           'Unnamed: 11' : 'T5killsDKP', 
                           'Deaths' : 'T4deadsDKP', 
                           'Unnamed: 13' : 'T5deadsDKP',
                           'DKP minus 20% of Base Power' : 'MinusBasePowerDKP',
                           'Power Pre-KVK' : 'PowerPreKVK', 
                           'Honor DKP' : 'HonorDKP',
                           'DKP Goal' : 'GoalDKP',
                           'Total DKP  (Kills, Deaths, Honor)' : 'TotalDKP'}, inplace = True) #renaming columns
renaming.head(5)
```

```python
renaming.insert(loc = 0, column = 'Ranking', value = [i for i in range(1, 240)]) #inserting "ranking" column
```

```python
renaming.head(5)
```

```python
power_vs_dkp = renaming
power_vs_dkp = power_vs_dkp.loc[:, ['PowerPreKVK', 'MinusBasePowerDKP']] #Selecting every row but only those 2 columns
print(power_vs_dkp.head(5))
DKP_avg = power_vs_dkp.MinusBasePowerDKP.mean() #Average dkp after - 20%
PreKVK_pow_avg = power_vs_dkp.PowerPreKVK.mean() #Average Pre-KVK power
print('AVG DKP = ', DKP_avg)
print('AVG Power = ', PreKVK_pow_avg)

```

```python
x = power_vs_dkp.PowerPreKVK
y = power_vs_dkp.MinusBasePowerDKP
```

```python
plt.scatter(x,y, color = 'red')
plt.xlabel('Pre-KVK Power')
plt.ylabel('DKP - 20% (Tens of millions)')
plt.title('Pre-KVK power VS DKP')
```

```python
below_50mil = (power_vs_dkp['PowerPreKVK'] <= 50000000) & (power_vs_dkp['MinusBasePowerDKP'] > 3000000)
subset_pwr_dkp = power_vs_dkp[below_50mil]
subset_pwr_dkp 
```
